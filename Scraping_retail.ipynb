```python
urls = [
    'https://www.frest.cl/producto/38-pimiento-rojo-un',
    'https://www.frest.cl/producto/33-papas-1-kg',
    'https://www.frest.cl/producto/59-limon-1-kg',
    'https://www.frest.cl/producto/2152-tomate-2a-seleccion-1-kg',
    'https://www.frest.cl/producto/30-lechuga-hidroponica-espanola-un',
    'https://www.frest.cl/producto/60-manzana-fuji-1-kg',
    'https://www.frest.cl/producto/377-platano-maduro-1-kg',
    'https://www.frest.cl/producto/49-zanahoria-1-kg',
    'https://www.frest.cl/producto/1900-cebolla-1-kg',
    'https://www.frest.cl/producto/50-zapallo-trozo',
    'https://www.frest.cl/producto/19-cebollin-3-un',
    'https://www.frest.cl/producto/994-palta-hass-chilena-1-kg',
    'https://www.frest.cl/producto/66-naranja-1-kg'

    ]
urls
```

```python
texto = []
for url in urls:
  page = requests.get(url)
  bs_page = BeautifulSoup(page.text, "html.parser")
  #print(page)
  for i in bs_page.find("span", {"class": "text-gray-600 line-through"}).children:
    precio = None
    nombre = None
    if precio is None:
      precio = i.text.strip()
    if nombre is None:
      nombre = url
    if precio is not None and nombre is not None:
      texto.append((nombre, precio))
      precio = None
      nombre = None
      break
    #print(url)
    #print(precio)
```

```python
texto
```

```python
pd_qas = pd.DataFrame(texto, columns=['nombre', 'precio'])
pd_qas['precio'] = pd_qas['precio'].str.strip('$')
pd_qas['precio'] = pd_qas['precio'].str.replace('.', '')
pd_qas['precio'] = pd_qas['precio'].astype(str).astype(int)
pd_qas
```

```python
print('Total Canasta Frest', pd_qas['precio'].sum())
```
