```python
from bs4 import BeautifulSoup
import requests
import urllib.parse
import pandas as pd
import random
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
from google.colab import drive
from datetime import datetime
```

```python
#Funciones
def crapjumbo(links):
  li = []
  for url in links:
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    for href in soup.find_all('a',href=True):
        link= None
        if link is None:
            link = href.get('href')
        if link is not None:
            if link.startswith("https://"):
                link=""
            elif link.startswith("/informaciones"):
                link=""
            elif link.startswith("#jumbo"):
                link=""
            elif link.startswith("/mis"):
                link=""
            elif link.startswith("/limpieza"):
                link=""
            elif link.startswith("/locales"):
                link=""
            elif link.startswith("/prime"):
                link=""
            else:
                li.append(("https://www.jumbo.cl"+link))
  li = list(set(li))
  return li
```

```python
def crapjumbo2(urls):
  jumbo = []
  for url in urls:
      page = requests.get(url)
      bs_page = BeautifulSoup(page.text, "html.parser")
      soup1 = bs_page.find("meta", property="product:price:amount")
      soup2 = bs_page.find("meta", property="product:brand")
      soup3 = bs_page.find("meta", property="twitter:title")
      soup4 = bs_page.find("meta", property="product:retailer_item_id")

      for i in bs_page.find_all("meta", property="product:price:amount"):
          precio = None
          nombre = None
          marca = None
          detalle = None
          item_retail = None

          if precio is None:
              precio = soup1["content"]
              #print(precio)
          if nombre is None:
              nombre = url
              #print(nombre)
          if marca is None:
              marca = soup2["content"]
              #print(marca)
          if detalle is None:
              detalle = soup3["content"]
              #print(detalle)
          if item_retail is None:
              item_retail = soup4["content"]
              #print(item_retail)

          if precio is not None and nombre is not None and marca is not None and detalle is not None and item_retail is not None:
              jumbo.append((nombre, precio, marca, detalle, item_retail))
              precio = None
              nombre = None
              marca = None
              detalle = None
              item_retail = None
              break
  return jumbo
```

```python
jumbo = []

for url in urls:
    page = requests.get(url)
    bs_page = BeautifulSoup(page.text, "html.parser")
    soup1 = bs_page.find("meta", property="product:price:amount")
    soup2 = bs_page.find("meta", property="product:brand")
    soup3 = bs_page.find("meta", property="twitter:title")
    soup4 = bs_page.find("meta", property="product:retailer_item_id")

    for i in bs_page.find_all("meta", property="product:price:amount"):
        precio = None
        nombre = None
        marca = None
        detalle = None
        item_retail = None

        if precio is None:
            precio = soup1["content"]
            #print(precio)
        if nombre is None:
            nombre = url
            #print(nombre)
        if marca is None:
            marca = soup2["content"]
            #print(marca)
        if detalle is None:
            detalle = soup3["content"]
            #print(detalle)
        if item_retail is None:
            item_retail = soup4["content"]
            #print(item_retail)

        if precio is not None and nombre is not None and marca is not None and detalle is not None and item_retail is not None:
            jumbo.append((nombre, precio, marca, detalle, item_retail))
            precio = None
            nombre = None
            marca = None
            detalle = None
            item_retail = None
            break
        #print(url)
        #print(precio)
```

```python
pd_qas2 = pd.DataFrame(jumbo, columns=['nombre', 'precio', 'marca', 'detalle', 'item_retail'])
pd_qas2["precio"] = pd_qas2["precio"].astype(str).astype(int)
pd_qas2["fecha"] = datetime.now()
pd_qas2["cat"] = 'detergente liquido'
pd_qas2
```

```python
fecha=datetime.now()
year_month = fecha.strftime('%Y-%m-%d')
archivo='data.csv'
#archivo_con_fecha = f"{fecha}_{archivo}"
pd_qas2.to_csv(f"{year_month}_{archivo}", index=False)
drive.mount('/content/drive', force_remount=True)
```
